### Load necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic graph

```python
x = [0,1,2,3,4]
y=[0,2,4,6,8]
#Resize the graph(dpi specifies pixels per inch. when saving probably should use 300)
plt.figure(figsize= (5,3))

#line1
#plt.plot([1,2,3],[2,4,6])
#keyword atgument notation
#plt.plot(x,y, label='2x', color='y', linewidth=2, marker='.', linestyle='--', markersize=10,markeredgecolor='blue')  #change color of line

#shorthand notation
#fnt='[color][marker][line]'
plt.plot(x,y,'r.--', label='2x')

#line no.2
x2 = np.arange(0,4,0.5)
#print(x2)
#line2
#select interval wanted to plot 
plt.plot(x2[:6],x2[:6]**2, 'r',label='X^')

#remainder plot of graph
plt.plot(x2[5:],x2[5:]**2,'r--')

#add title -specify fornt parameters with fontdict
plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

#x and y labels
plt.xlabel('X Axis', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y Axis')

#x , y axis tickmarks scale of the graph
plt.xticks([0,1,2,3,4])  #resize the graph and turn int
#plt.yticks([0,2,4,6,8,10])
plt.legend()

#save figure (dpi 300 is good to get high resolution)
plt.savefig('mygraph.png', dpi=300)   #save the graph

#show plot
plt.show()
```

### Bar chart

```python
labels= ['A','B','C']
values= [1,4,2]
bars = plt.bar(labels, values)
patterns= ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/') #shape with /
#bars[1].set_hatch('*')
#bars[2].set_hatch('0')
plt.figure(figsize=(6,4))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashdot', c='#ff00ff', linewidth='2')  # :=dotted,--=dashed,-.=dashdot , or=none
plt.title("Dotted Graph", loc = 'left')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(color = 'green', linestyle = '-.', linewidth = 0.5)
plt.show()
```
